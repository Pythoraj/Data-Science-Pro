## 2. Describe the key components of LeNet-5 and their roles in the network

---
**Key components of LeNet-5 and their roles in the network:**

1. **Input Layer**:
   - **Role**: Accepts 32x32 grayscale images.
   - **Purpose**: Prepares the image for further processing by the network.

2. **Convolutional Layer (C1)**:
   - **Role**: Consists of 6 filters, each 5x5 in size, resulting in 6 feature maps of size 28x28.
   - **Purpose**: Detects local features like edges and corners from the input image.

3. **Subsampling Layer (S2)**:
   - **Role**: Performs average pooling with a 2x2 filter, reducing each feature map to 14x14.
   - **Purpose**: Reduces the dimensionality and computational load while retaining important features.

4. **Convolutional Layer (C3)**:
   - **Role**: Uses 16 filters of size 5x5, producing 16 feature maps of size 10x10.
   - **Purpose**: Captures more complex patterns and features from the subsampled maps. 

5. **Subsampling Layer (S4)**:
   - **Role**: Another average pooling layer with 2x2 filters, reducing feature maps to 5x5.
   - **Purpose**: Further reduces the size of the feature maps while preserving significant information.

6. **Convolutional Layer (C5)**:
   - **Role**: Applies 120 filters of size 5x5, resulting in 120 feature maps of size 1x1.
   - **Purpose**: Fully connects each feature map to a set of neurons, capturing high-level representations.

7. **Fully Connected Layer (F6)**:
   - **Role**: Contains 84 neurons.
   - **Purpose**: Connects all neurons from the previous layer and performs further abstraction to prepare for output.

8. **Output Layer**:
   - **Role**: Consists of 10 neurons (one for each digit from 0 to 9) using the softmax activation function.
   - **Purpose**: Produces the final classification result, representing the probability distribution over the 10 classes.

Each component plays a crucial role in transforming the input image through successive layers of feature extraction and abstraction, ultimately leading to accurate classification.

## 3. Discuss the limitations of LeNet-5 and how subsequent architectures like AlexNet addressed these limitations.

### Limitations of LeNet-5:
---
1. **Limited Depth**:
   - LeNet-5 had only seven layers, which limited its ability to capture complex features and patterns.
  
2. **Small Dataset Suitability**:
   - It was primarily designed for small, grayscale images (like handwritten digits) and did not scale well to larger, more complex datasets with color images.

3. **Simple Activation Functions**:
   - LeNet-5 used sigmoid and tanh activation functions, which can suffer from vanishing gradient problems, making it harder to train deep networks.

4. **Computational Constraints**:
   - At the time, computing power was limited, which restricted the size and depth of the network that could be effectively trained.
---
### How AlexNet Addressed These Limitations:

1. **Increased Depth and Size**:
   - **AlexNet** has eight layers, including five convolutional layers and three fully connected layers, allowing it to learn more complex and abstract features.

2. **Use of ReLU Activation**:
   - ReLU (Rectified Linear Unit) activation function was introduced in AlexNet, addressing the vanishing gradient problem and enabling faster training.

3. **Dropout for Regularization**:
   - AlexNet introduced dropout layers to prevent overfitting, making the network more robust when training on large datasets.

4. **GPU Utilization**:
   - By leveraging GPUs for training, AlexNet significantly reduced training time and allowed for the processing of larger networks and datasets.

5. **Data Augmentation and Normalization**:
   - Techniques like data augmentation and normalization were used to improve the network's performance and robustness.

6. **Larger Filters and Feature Maps**:
   - AlexNet used larger filters (e.g., 11x11, 5x5) in the initial layers and more feature maps, which helped in capturing more detailed and varied features from the input images.

By addressing these limitations, AlexNet was able to achieve remarkable success in image recognition tasks, famously winning the ImageNet Large Scale Visual Recognition Challenge (ILSVRC) in 2012. This victory marked a significant turning point in the adoption and development of deep learning technologies.

## 4. Explain the architecture of AlexNet and its contributions to the advancement of deep learning.

## The AlexNet Architecture:

---


AlexNet comprises a rather simple architecture compared to the latest Deep Learning Models. It consists of 8 layers: 5 convolutional layers and 3 fully connected layers.

However, it integrates several key innovations of its time, including the ReLU Activation Functions, Local Response Normalization (LRN), and Overlapping Max Pooling. We will look at each of them below.

**Input Layer**
AlexNet takes images of the Input size of 227x227x3 RGB Pixels.

**Convolutional Layers**
*First Layer:* The first layer uses 96 kernels of size 11×11 with a stride of 4, activates them with the ReLU activation function, and then performs a Max Pooling operation.
- *Second Layer:* The second layer takes the output of the first layer as the input, with 256 kernels of size 5x5x48.
- *Third Layer:* 384 kernels of size 3x3x256. No pooling or normalization operations are performed on the third, fourth, and fifth layers.
- *Fourth Layer:* 384 kernels of size 3x3x192.
- *Fifth Layer:* 256 kernels of size 3x3x192.

**Fully Connected Layers**
The fully connected layers have 4096 neurons each.

**Output Layer**
The output layer is a SoftMax layer that outputs probabilities of the 1000 class labels.

---
## Stochastic Gradient Descent

AlexNet uses stochastic gradient descent (SGD) with momentum a for optimization. SGD is a variant of Gradient Descent. In SGD, instead of performing calculations on the whole dataset, which is computationally expensive, SGD randomly picks a batch of images, calculates loss, and updates the weight parameters.


**Momentum**
Momentum helps accelerate SGD and dampens fluctuations. It performs this by adding a fraction of the previous weight vector to the current weight vector. This prevents sharp updates and helps the model overcome saddle points.

**Weight Decay**
Lnew(w) = Loriginal(w) + λwTw

Weight decay adds a penalty on the size of the weights. By penalizing larger weights, weight decay encourages the model to learn smaller weights that capture the underlying patterns in the data, rather than memorizing specific details that results in overfitting.

**Data Augmentation**
Data Augmentation artificially increases the size and diversity of the training data using transformations that don’t change the actual object in the image. This forces the model to learn better and increase its ability to cater to unseen images.

AlexNet used Random Cropping and Flipping, and Color Jitter Augmentations.

## Innovations Introduced

### ReLU:


This is a non-linear activation function that has a simple derivative. (It eradicates negative values and only outputs positive values).

Before ReLU, Sigmoid and Tanh functions were used, but they slowed down the training and caused gradient vanishing. ReLU came up as an alternative since it is faster and requires fewer computation resources during backpropagation because the derivative of Tanh is always less than 1, while the derivative of ReLU is either 0 or 1.

This is a regularization technique (a technique used to address overfitting) that “drops out” or turns off neurons.

This is a simple method that prevents the model from overfitting. It does this by making the model learn more robust features and stopping the model from condensing features in a single area of the network.

Dropout increases the training time; however, the model learns to generalize better.

### Overlapping Pooling
Pooling layers in CNNs summarize the outputs of neighboring groups of neurons in the same kernel map. Traditionally, the neighborhoods summarized by adjacent pooling units do not overlap. However, in Overlapping pooling the adjacent pooling operations overlap with each other. This approach:

- Reduces the size of the network
- Provides slight translational invariance.
- Makes it difficult for the model to overfit.

### Performance and Impact
The AlexNet architecture dominated in 2012 by achieving a top-5 error rate of 15.3%, significantly lower than the runner-up’s 26.2%.

This large reduction in error rate excited the researchers with the untapped potential of deep neural networks in handling large image datasets. Subsequently, various Deep Learning models were developed later.

---
### AlexNet inspired Networks
The success of AlexNet inspired the design and development of various neural network architectures. These include:

- *VGGNet:* It was developed by K. Simonyan and A. Zisserman at Oxford. The VGGNet incorporated ideas from AlexNet and was the next step taken after AlexNet.
- *GoogLeNet (Inception):* This was Introduced by Google, which further developed the architecture of AlexNet.
- *ResNets:* AlexNet started to face a vanishing gradient with deeper networks. To overcome this, ResNet was developed. These networks introduced residual learning, also called skip connections. The skip connection connects activations of lower layers to higher layers by skipping some layers in between.

### Applications of AlexNet

Developers created AlexNet for image classification. However, advances in its architecture and transfer learning (a technique where a model trained on one task is repurposed for a novel related task) opened up a new set of possibilities for AlexNet. Moreover, its convolutional layers form the foundation for object detection models such as Fast R-CNN and Faster R-CNN, and professionals have utilized them in fields like autonomous driving and surveillance.

### Compare and contrast the architectures of LeNet-5 and AlexNet.
---



### Limitations of LeNet-5:
1. **Limited Depth**:
   - LeNet-5 has a relatively shallow architecture with fewer layers, limiting its ability to capture complex hierarchical patterns in data.

2. **Training Data and Overfitting**:
   - It was designed primarily for small datasets like MNIST (handwritten digits), and it struggled with larger, more complex datasets.

3. **Activation Functions**:
   - LeNet-5 used sigmoid and tanh activation functions, which can suffer from vanishing gradients, slowing down training and reducing effectiveness for deeper networks.

4. **Computational Resources**:
   - During the time LeNet-5 was introduced, the computational power was limited, which restricted the depth and complexity of the network.

### How AlexNet Addressed These Limitations:
1. **Increased Depth and Complexity**:
   - AlexNet is much deeper with eight learned layers, including five convolutional layers and three fully connected layers. This depth allows for the extraction of more complex features and patterns.

2. **ReLU Activation Function**:
   - AlexNet introduced the ReLU (Rectified Linear Unit) activation function, which mitigates the vanishing gradient problem and speeds up training.

3. **GPU Utilization**:
   - AlexNet leveraged GPUs for training, significantly reducing training time and enabling the training of much larger models on large datasets.

4. **Data Augmentation and Dropout**:
   - To combat overfitting, AlexNet used data augmentation (techniques like cropping, flipping, and rotating images) and dropout regularization, which randomly drops units during training to prevent co-adaptation.

5. **Large-Scale Dataset Training**:
   - AlexNet was trained on the ImageNet dataset, containing millions of images with thousands of classes, demonstrating the network's ability to generalize to a wide variety of data.

6. **Larger Filters and Stride in Initial Layers**:
   - The initial layers of AlexNet use larger filters (e.g., 11x11), and strides to capture more spatial information. This is particularly useful for recognizing complex, high-dimensional patterns in images.
