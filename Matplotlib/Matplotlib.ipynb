```python
plt.figure()           # It will create a new figure or exicuted existing figure
plt.subplot(3,3,1)     # Here we have divided a plot into 3x3 formate where first three is rows and second 3 is columns. 1 indicated as index
plt.plot(x, np.sin(x), '|', color='r') # this is is the main plot

plt.subplot(3,3,2)     # 3-rows, 3-columns and index position is 2
plt.plot(x, np.cos(x), color = 'g')

plt.subplot(3,3,3)     # 3-rows, 3-columns and index position is 3 
plt.plot(x, np.tan(x), color = 'y')

plt.subplot(3,3,4)     # 3-rows, 3-columns and index position is 4
plt.plot(x, np.sin(x+1), color = 'b')

plt.subplot(3,3,5)     # 3-rows, 3-columns and index position is 5
plt.plot(x, np.cos(x+1), color = 'w')

plt.subplot(3,3,6)     # 3-rows, 3-columns and index position is 6
plt.plot(x, np.tan(x+1), color = 'm')

plt.subplot(3,3,7)     # 3-rows, 3-columns and index position is 7
plt.plot(x, np.sin(x+2), color = 'c')

plt.subplot(3,3,8)     # 3-rows, 3-columns and index position is 8
plt.plot(x, np.cos(x+2), color = '#d703fc')

plt.subplot(3,3,9)     # 3-rows, 3-columns and index position is 9
plt.plot(x, np.tan(x+2), color = '#fcad03')

plt.show()
```

```python
# Column-wise data representation in plot

plt.figure()
plt.subplot(2,1,1)
plt.plot(x, np.sin(x), scalex=True, scaley=True)

plt.subplot(2,1,2)
plt.plot(x,np.cos(x),scalex=True, scaley=True)

plt.show()
```

```python
# scale the plot size

f = plt.figure()

f.set_figwidth(4)
f.set_figheight(1)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x), scalex=True, scaley=True)

plt.subplot(2,1,2)
plt.plot(x,np.cos(x),scalex=True, scaley=True)

plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x), color = 'b')
plt.xlim(1, 15)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.axis('equal')
plt.title("This is sin plot before real dataset execution to understand basic syntax", color = 'r',loc='center')

plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x), color = 'g')
plt.xlim(1,15)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.axis('equal')
plt.title('This is sin plot before real dataset execution to understand basic syntax', color = "y")
plt.xlabel("X-Axis", color = 'b')
plt.ylabel("Y-Axis", color = 'b')
plt.show()
```

```python
plt.figure()
plt.plot(x, np.sin(x-0), color = 'g', label = 'X-Axis')
plt.plot(x, np.cos(x-0), color = 'y', label = 'Y-Axis')
plt.xlim(1,15)
plt.ylim(-1.5,1.5)
plt.xlabel('X-Axis', color = 'y')
plt.ylabel('Y-Axis', color = 'y')
plt.grid(color = 'r')
plt.axis('equal')
plt.title("This is sin plot before real dataset execution to understand basic syntax", color = 'b')
plt.legend()
plt.show()
```
